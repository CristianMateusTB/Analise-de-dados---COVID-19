```python
#Importando as bibliotecas:
import pandas as pd
from matplotlib import pyplot as plt
low_memory=False
%matplotlib inline
```

```python
filename=r'INFLUD-30-06-2020.csv' #Informações retiradas desse banco de dados
df=pd.read_csv(filename,delimiter=';',
              encoding='ISO-8859-1',
              error_bad_lines=False)
     
```

```python
df.shape
```

```python
df.info()
```

```python
df.groupby('ID_MUNICIP').size().sort_values().tail(7).plot(kind='bar',color= '#AB2D12',edgecolor='black')
plt.style.use("ggplot")
plt.rcParams['figure.figsize'] = (21, 21) #Tamanho
plt.title('Cidades com maiores casos',fontsize=22 ) #Título do gráfico
plt.xticks(rotation=45) # Rotacionar os nomes das cidades
plt.xlabel('')
plt.show()
```

```python
df.ID_MUNICIP.value_counts().head(7).to_frame('Cidades')
```

```python
df.groupby('CS_SEXO').size().sort_values().tail(2).plot(kind='barh', color= '#1F7056',edgecolor='black') #Quantidades de colunas(M,F),Cor e tipo do gráfico
plt.rcParams['figure.figsize'] = (5, 5)
plt.title('Análise de casos por sexo',fontsize=22)#Tamanho
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')#Rotacionar as legendas
plt.ylabel('')#Deixar label do eixo y vazia
plt.show()
```

```python
df.CS_SEXO.value_counts().head(2).to_frame('Sexo')
```

```python
df.groupby('NU_IDADE_N').size().plot(kind='line').set_title('Análise de casos por idade', fontsize = 21)#Tipo de gráfico e tamanho
plt.xlabel('')# Deixar label do eixo x vazia
plt.show()
```

```python
df.NU_IDADE_N.value_counts().head(10).to_frame('Idade')
```

```python
df.groupby('CS_ETINIA').size().sort_values().tail(7).plot(kind='bar',color= '#7E3005',edgecolor='black')# 7 grupos de etnias, tipo de gráfico e cor.
plt.rcParams['figure.figsize'] = (21, 21) #Tamanho
plt.title('Casos nas principais etnias indígenas',fontsize=22 ) #Título do gráfico
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right') # Rotacionar os nomes das etnias
plt.xlabel('')# Deixar label x vazia
plt.show()
```

```python
df.CS_ETINIA.value_counts().head(7).to_frame('Etinia')
```
